```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = '/raid/asap7772/epic100/epic100_bridgeform/epic100/sponge-pick-up/train/out.npy'
# path = '/raid/trainingdata/robonetv2/toykitchen_fixed_cam_targetdom_anikait_numpy/toykitchen6/croissant_in_pan/train/out.npy'
```

```python
data = np.load(path, allow_pickle=True)
print(len(data))
print(len(data[0]['observations']))
```

```python
num_rows = 3
num_cols = (len(data[0]['observations']) // num_rows) + 1
print(num_rows, num_cols)

i = 0
fig, axs = plt.subplots(num_rows, num_cols, figsize=(10, 5))

for x in data[0]['observations']:
    which_row = i // num_cols
    which_col = i % num_cols
    
    axs[which_row][which_col].imshow(x['images0'])
    i += 1

for i in range(num_rows):
    for j in range(num_cols):
        axs[i][j].axis('off')
        axs[i][j].grid(b=None)

fig.tight_layout()    
plt.show()
```
